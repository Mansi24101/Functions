6. Explain the concept of generators in Python and how they are defined.
- A generator is a special type of iterator.
It produces values one at a time using the yield keyword instead of return.
Generators do not store all values in memory at once.
The generator can be defined using 2 ways: using a function with yield and using the generator expression like list comprehension but with parentheses() instead of[]


```python
#generator
def my_generator():
    yield 1
    yield 2
    yield 3
generator = my_generator()
for value in generator:
    print(value)
```

```python
#generator expression
my_list = [1, 2, 3, 4, 5, 6, 7]
generator = (x**2 for x in my_list)
for value in generator:
    print(value)
```

4.  Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return False
    return True

for number in range(1, 201):
    if is_prime(number):
        print(number, "is a prime number")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, max_terms):
        self.max_terms = max_terms
        self.count = 0
        self.current, self.next_value = 0, 1

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.max_terms:
            raise StopIteration
        current_term = self.current
        self.current, self.next_value = self.next_value, self.current + self.next_value
        self.count += 1
        return current_term

fib = FibonacciIterator(10)
for number in fib:
    print(number)
```

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(exponent):
    for i in range(exponent + 1):
        yield 2 ** i

for power in powers_of_two(5):
    print(power)
```

 7. Implement a generator function that reads a file line by line and yields each line as a string

```python
file_name = 'example.txt'

with open(file_name, 'w') as file:
    file.write("Hello world\n")
    file.write("This is Mansi\n")
    file.write("I love travelling\n")
    file.write("Travelling is fun\n")
    file.write("Keep going\n")

def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.strip()
file_path = 'example.txt'
lines = read_file_line_by_line(file_path)

print("Output:")
for line in lines:
    print(line)
```

```python
celsius_temperatures = [0, 10, 20, 30, 40, 50]
def celsius_to_fahrenheit(celsius):
    return (celsius * 9/5) + 32

fahrenheit_temperatures = list(map(celsius_to_fahrenheit, celsius_temperatures))
print(fahrenheit_temperatures)
```

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

```python
def is_not_vowel(char):
    return char.lower() not in 'aeiou'

input_string = "Hello World!"
filtered_string = ''.join(filter(is_not_vowel, input_string))
print(filtered_string)
```
